```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
import matplotlib as mpl

# 手动设置 Arial 的路径
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['axes.titlesize'] = 22  # 设置标题字体大小
mpl.rcParams['axes.labelsize'] = 15  # x轴和y轴标签字体大小
mpl.rcParams['xtick.labelsize'] = 22  # x轴刻度标签字体大小
mpl.rcParams['ytick.labelsize'] = 22  # y轴刻度标签字体大小
mpl.rcParams['legend.fontsize'] = 22  # y轴刻度标签字体大小
mpl.rcParams['axes.titlesize'] = 22  # 设置标题大小为 16
fig, axs = plt.subplots(1, 3, figsize=(16,5))
colors = [
    (187/255, 187/255, 187/255),  # 灰色
    (254/255, 218/255, 185/255),  # 橙色
    (159/255, 187/255, 224/255),  # 蓝色
    (254/255, 132/255, 196/255),  # 粉色
    (189/255, 228/255, 189/255),  # 绿色
]
```

```python
# 成员A定义
categories = ['Ex-ante Stage', 'Ex-post Stage']
group1 = [14.88,-12.00]  # Individual
group2 = [15.32, -10.33]  # Shapley
group3 = [15.33, -10.17]  # Equal DP
group4 = [15.33, -10.21]   # Two-Stage
group5=[0,-19.875]      # Shadow Imputation 
group6=[0,-10.46]     # Expectation Allocation
bar_width = 0.15
x = np.arange(len(categories))

# 创建分组柱状图
axs[0].bar(x - bar_width, group1, width=bar_width, label='Individual', edgecolor='black', linewidth=1,color=colors[0])
axs[0].bar(x, group2, width=bar_width, label='Shapley', edgecolor='black', linewidth=1,color=colors[1])
axs[0].bar(x + bar_width, group3, width=bar_width, label='Equal DP', edgecolor='black', linewidth=1,color=colors[2])
axs[0].bar(x + 2 * bar_width, group4, width=bar_width, label='Corrected Equal DP', edgecolor='black', linewidth=1,color=colors[3])
axs[0].bar(x + 3 * bar_width, group5, width=bar_width, label='Shadow Imputation\n(Pure Ex-post)', edgecolor='black', linewidth=1,color=colors[4])

# 添加虚线
extreme_value_individual = group1[1]  # Extreme 组中 Individual 的值
axs[0].plot([x[1] - bar_width - bar_width / 2, x[1] + 3 * bar_width + bar_width / 2],
         [extreme_value_individual, extreme_value_individual], 
         linestyle='--', color='red', linewidth=2)
extreme_value_individual_1 = group1[0]  # Extreme 组中 Individual 的值
axs[0].plot([x[0] - bar_width - bar_width / 2, x[0] + 2* bar_width + bar_width / 2],
         [extreme_value_individual_1, extreme_value_individual_1], 
         linestyle='--', color='red', linewidth=2)
axs[0].axhline(y=0, color='black', linestyle='-', linewidth=1)

# 分配虚线
Expection_Line = group6[1]  # Extreme 组中 Individual 的值
axs[0].plot([x[1]+ bar_width / 2 , x[1]+2*bar_width + bar_width / 2],
         [Expection_Line, Expection_Line], 
         linestyle='--', color='green', linewidth=2)
# 添加标签和标题
axs[0].set_ylabel('Profit($)',fontname='Arial')
axs[0].set_title('Member A',fontname='Arial')
axs[0].set_xticks([0.1, 1.2])  # 向右偏移
axs[0].set_xticklabels(['Ex-ante Stage', 'Ex-post Stage'],fontname='Arial',fontsize=19.5)
#axs[0].set_xticks(x, categories)
axs[0].set_yticks(np.arange(-20,20,5))
```
